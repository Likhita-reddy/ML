```python
df= pd.read_csv("/kaggle/input/sentiment-analysis-on-movie-reviews/sampleSubmission.csv")
df['PhraseId'].to_csv('phrase')
```

```python
train = pd.read_csv("/kaggle/input/sentiment-analysis-on-movie-reviews/train.tsv.zip",sep="\t")
test = pd.read_csv("/kaggle/input/sentiment-analysis-on-movie-reviews/test.tsv.zip",sep='\t')
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer(stop_words='english',lowercase=True,ngram_range=(1,2),max_features=10000)
final_train = vectorizer.fit_transform(train['Phrase'])
final_test = vectorizer.fit_transform(test['Phrase'])
```

```python
final_train=final_train.toarray().astype(float)
```

```python
max(final_train[10])
```

```python
import tensorflow as tf

tf.random.set_seed(42)
```

```python
model = tf.keras.Sequential()
model.add(tf.keras.layers.InputLayer(input_shape=final_train.shape[1]))
model.add(tf.keras.layers.Dense(200,activation='relu'))
model.add(tf.keras.layers.Dense(100,activation='relu'))
model.add(tf.keras.layers.Dense(10,activation='relu'))
model.add(tf.keras.layers.Dense(5,activation='softmax'))
```

```python
model.compile(loss='sparse_categorical_crossentropy',optimizer='sgd',metrics=['accuracy'])
```

```python
callback = tf.keras.callbacks.EarlyStopping(monitor='val_loss',min_delta=0.0001,patience=3)
```

```python
model.fit(final_train,train['Sentiment'],epochs=10,callbacks=callback)
```

```python
pred_test=model.predict(final_test.toarray())
```

```python
pred_test
```

```python
pred_test = pred_test.argmax(axis=-1)
pred_test
```

```python
data = pd.DataFrame(pred_test)
```

```python
data.to_csv('Submission.csv')
```
